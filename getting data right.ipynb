```python
def Bat_data(player,matchID):
    dic= {
    "overall":0,
     'overall_F6':0,
    "flat":0,
    "slow":0,
    'Medium':0,   
    'Right Fast':0,
    'Off break':0,
    'Left Wrist':0,
    'Left Spin':0,
    'Left Fast':0,
    'Right Wrist':0,

    "last5_SR":0,
    'average':0,
    'last5_avg':0
    }
    zols = {
    'flat' : 0,
    'slow' : 0,
    'Medium':0,   
    'Right Fast':0,
    'Off break':0,
    'Left Wrist':0,
    'Left Spin':0,
    'Left Fast':0,
    'Right Wrist':0
    }
    playerD = ndata[ndata['striker']==player]#.reset_index()
    if len(playerD[playerD['match_id']<matchID])!=0:
        playerD = playerD[playerD['match_id']<matchID].reset_index()   #####################debue
    else:
        playerD = playerD.reset_index()
        print('debdeb')
    playerD = playerD[playerD['start_date']>=playerD['start_date'][len(playerD)-1]-3 ]#.reset_index()

    dic['overall'] = playerD['runs_off_bat'].sum()/len(playerD)
    
    if len(playerD[playerD['ball']<6.0])!=0:
        playerD=playerD[playerD['ball']<6.0].reset_index()
    else:
        playerD=playerD.reset_index()
    
    dic['overall_F6'] = playerD['runs_off_bat'].sum()/len(playerD)

    for i in range(len(playerD)):
    
        dic[ pitch['pitch_type'][ playerD['venue'][i] ] ]+=playerD['runs_off_bat'][i]
        zols[ pitch['pitch_type'][ playerD['venue'][i] ] ]+=1

        dic[ skill['bowl'][playerD['bowler'][i] ]] +=playerD['runs_off_bat'][i]
        zols[ skill['bowl'][playerD['bowler'][i] ] ]+=1
    
    dic['average']=playerD['runs_off_bat'].sum()/playerD['match_id'].nunique()
    
    last_m = playerD['match_id'].unique()
    count=1
    last5_sr=0
    last5_avg=0
    for i in range(len(last_m)-1,-1,-1):
        
        temp = playerD[playerD['match_id']==last_m[i]]
        
        last5_sr += temp['runs_off_bat'].sum()/len(temp)
        
        last5_avg += temp['runs_off_bat'].sum()
        if count >6: #5 
            break
        count+=1
    
    dic['last5_SR']=last5_sr/(count)
    dic['last5_avg']=last5_avg/(count)
    
    #print(dic)
    return dic , zols 
Bat_data("TM Head",1254060)
```

```python
def Bowl_data(player,matchID):
    dic={
    "overall_eco":0,
    "flat":0,
    "slow":0,
    "last5_eco":0,
    'last5_wickets':0
    }
    zols = {
    'flat' : 0,
    'slow' : 0
    }
    playerD = ndata[ndata['bowler']==player]#.reset_index()
    
    if len(playerD[playerD['match_id']<matchID])!=0:
        playerD = playerD[playerD['match_id']<matchID].reset_index()   #####################debue
        
    else:
        playerD = playerD.reset_index()
        print('debu')
        
    playerD = playerD[playerD['start_date']>=playerD['start_date'][len(playerD)-1]-3 ]#.reset_index()
    
    dic['overall_eco'] = playerD['runs_off_bat'].sum()#/(len(playerD)-len(playerD)%6)*6
    #print(playerD)
    playerD=playerD[playerD['ball']<6.0].reset_index()
    
    for i in range(len(playerD)):
        dic[ pitch['pitch_type'][ playerD['venue'][i] ] ]+=playerD['runs_off_bat'][i]
        zols[ pitch['pitch_type'][ playerD['venue'][i] ] ]+=1
        
    last_m = playerD['match_id'].unique()
    count=1
    last5_eco=0
    last5_wickets=0
    #print(last_m)
    for i in range(len(last_m)-1,-1,-1):
        
        temp = playerD[playerD['match_id']==last_m[i]]
        
        if (len(temp)-len(temp)%6) != 0:
            last5_eco += temp['runs_off_bat'].sum()/(len(temp)-len(temp)%6)*6

            last5_wickets += temp['wicket_type'].nunique()
            if count >6: #5 
                break
            count+=1
    dic['last5_wickets']=last5_wickets
    dic['last5_eco']=last5_eco/count
    return dic , zols
Bowl_data('Rashid Khan', 1082591.0)
```

```python
def Venue_data(ven,matchID):
    v_d = ndata[ndata['venue']==ven]
    v_d = v_d[v_d['match_id']<matchID]
    
    v_d = v_d[v_d['ball']<6.0]
    last_m = v_d['match_id'].unique()
    count=1
    summ=0
    for i in range(len(last_m)-1,-1,-1):
        
        match = v_d[v_d['match_id']==last_m[i]]
        
        summ += match[match['innings']==1]["runs_off_bat"].sum() + match[match['innings']==1]["extras"].sum()
        summ += match[match['innings']==2]["runs_off_bat"].sum() + match[match['innings']==2]["extras"].sum()
        if count >6: #5 
            break
        count+=1
    return summ/(count*2)
print(Venue_data('MA Chidambaram Stadium',1254060))
```

```python
def Team_data(team,pitchType,matchID):
    d = ndata[ndata['batting_team']==team]
    d = d[d['ball']<6.0]
    if len(d[d['match_id']<matchID])!=0:
        d = d[d['match_id']<matchID]
    else:
        print(team,pitchType,matchID)
    
    summ=0
    c=1
    mat = d['match_id'].unique()
    #print(mat)
    for i in range(len(mat)-1,0,-1):
        if pitch['pitch_type'][ d[d["match_id"]==mat[i]].reset_index()['venue'][0] ]==pitchType:
            summ+= d[d["match_id"]==mat[i]]['runs_off_bat'].sum() + d[d["match_id"]==mat[i]]['extras'].sum()
            #print(summ,c)
            c+=1
        if c>5:
            break
        
    return summ/c
    
Team_data('Sunrisers Hyderabad','slow',1254060)
```

```python
def Bat_W_A(lis,matchID):
    dic= {
    "overall":0,
     'overall_F6':0,
    "flat":0,
    "slow":0,
    'Medium':0,   
    'Right Fast':0,
    'Off break':0,
    'Left Wrist':0,
    'Left Spin':0,
    'Left Fast':0,
    'Right Wrist':0,

    "last5_SR":0,
    'average':0,
    'last5_avg':0
    }
    zol = {
    'flat' : 0,
    'slow' : 0,
    'Medium':0,   
    'Right Fast':0,
    'Off break':0,
    'Left Wrist':0,
    'Left Spin':0,
    'Left Fast':0,
    'Right Wrist':0
    }
    for i in lis:
        ddic , zols =Bat_data(i,matchID)
    #print(ddic)
        for k in dic.keys():
            dic[k]+=ddic[k]
        for k in zol.keys():
            zol[k] += zols[k]

    for k in zol.keys():
        if zol[k]!=0:
            dic[k]= dic[k]/zol[k]

    for i in set(dic.keys())-set(zol.keys()):
        dic[i] = dic[i]/len(lis)
    
    return dic

print(Bat_W_A(['DA Warner','MK Pandey'],1254060.0))
```

```python
def Bol_W_A(lis,matchID):
    dic={
    "overall_eco":0,
    "flat":0,
    "slow":0,
    "last5_eco":0,
    'last5_wickets':0
    }
    zols = {
    'flat' : 0,
    'slow' : 0
    }
    for p in lis:
        dicc , z = Bowl_data(p,matchID)
        for k in dic.keys():
            dic[k]+=dicc[k]
        for k in z.keys():
            zols[k] += z[k]
    #print(dicc,z)
    for k in zols.keys():
        if zols[k]!=0:
            dic[k]= dic[k]/zols[k]*6
  
    dic['overall_eco'] = dic['overall_eco'] /(zols['flat']+zols['slow'])*6/len(lis)
    if dic['overall_eco'] >100:
        print(lis,matchID)
    dic['last5_eco'] = dic['last5_eco'] /len(lis)
    #print(dic,zols)
    dip={
    "overall_eco":dic['overall_eco'],
    "flat_eco":dic['flat'],
    "slow_eco":dic['slow'],
    "last5_eco":dic['last5_eco'],
    'last5_wickets':dic['last5_wickets']
    }
    return dip

print(Bol_W_A(['AD Russell','M Prasidh Krishna','PJ Cummins'],1254060))
```
