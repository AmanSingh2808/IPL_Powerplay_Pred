```python
def Bat_data(player):
  dic= {
    "overall":0,
    "flat":0,
    "slow":0,
    "batting":0,
    #ballertype
    "last5":0,
    'average':0,
    'last5_avg':0
  }
   zols = {
  'flat' : 0,
  'slow' : 0,
  'batting':0
  }
  player_d = ndata[ndata['striker']==player].reset_index()
  #print(player_d)
 
  for i in range(len(player_d)):
    dic[ pitch['pitch_type'][ player_d['venue'][i] ] ]+=player_d['runs_off_bat'][i]
    zols[ pitch['pitch_type'][ player_d['venue'][i] ] ]+=1

  dic['overall'] = player_d['runs_off_bat'].sum()/len(player_d)

  dic['flat'] = dic['flat']/zols['flat']
  dic['slow'] = dic['slow']/zols['slow']
  dic['batting'] = dic['batting']/zols['batting']
  
  last_5_m = player_d['match_id'].unique()[-5:]
  print(last_5_m)
  last_5=0
  l5_average=0
  for i in last_5_m:
    temp = player_d[player_d['match_id']==i]
    last_5+=temp['runs_off_bat'].sum()/len(temp)
    l5_average=temp['runs_off_bat'].sum()
  dic['last5']=last_5/5
  dic['last5_avg']=l5_average/5

  dic['average']=player_d['runs_off_bat'].sum()/player_d['match_id'].nunique()

  return dic


print(Bat_data("MK Pandey"))
```

```python
def Bowl_data(player):
  dic={
      "overall":0,
       "flat":0,
    "slow":0,
    "batting":0,
    "last5":0
  }
  zols = {
  'flat' : 0,
  'slow' : 0,
  'batting':0
  }
  player_d = ndata[ndata['bowler']==player].reset_index()
  #print(player_d)
  for i in range(len(player_d)):
    dic[ pitch['pitch_type'][ player_d['venue'][i] ] ]+=player_d['runs_off_bat'][i]
    zols[ pitch['pitch_type'][ player_d['venue'][i] ] ]+=1

  dic['overall'] = player_d['runs_off_bat'].sum()/(len(player_d)-len(player_d)%6)*6

  dic['flat'] = dic['flat']/(zols['flat']-zols['flat']%6)*6
  dic['slow'] = dic['slow']/(zols['slow']-zols['slow']%6)*6
  dic['batting'] = dic['batting']/(zols['batting']-zols['batting']%6)*6
  
  last_5_m = player_d['match_id'].unique()[-5:]
  #print(last_5_m)
  last_5=0
  last_5_b=0
  for i in last_5_m:
    temp = player_d[player_d['match_id']==i]
    last_5+=temp['runs_off_bat'].sum()
    last_5_b+=len(temp) 
    #print(temp['runs_off_bat'].sum(),len(temp)- len(temp)%6)
  dic['last5']=last_5/(last_5_b-last_5_b%6)*6
  

  return dic


Bowl_data("AD Russell")
```

```python
def Venue_data(ven):
  v_d = ndata[ndata['venue']==ven]
  v_d = v_d[v_d['ball']<6.0]
  last_5 = v_d['match_id'].unique()[-5:]
  summ=0
  for i in last_5:
    match = v_d[v_d['match_id']==i]
    summ += match[match['innings']==1]["runs_off_bat"].sum() + match[match['innings']==1]["extras"].sum()
    summ += match[match['innings']==2]["runs_off_bat"].sum() + match[match['innings']==2]["extras"].sum()
    #print(summ)
  return summ/10
print(Venue_data('MA Chidambaram Stadium'))
```

```python
def Team_data(team,pitchType):
  d = ndata[ndata['batting_team']==team]
  d = d[d['ball']<6.0]
  summ=0
  c=0
  mat = d['match_id'].unique()
  for i in range(len(mat)-1,0,-1):
    if pitch['pitch_type'][ d[d["match_id"]==mat[i]].reset_index()['venue'][0] ]==pitchType:
      summ+= d[d["match_id"]==mat[i]]['runs_off_bat'].sum() + d[d["match_id"]==mat[i]]['extras'].sum()
      #print(d[d["match_id"]==mat[i]]['runs_off_bat'].sum() + d[d["match_id"]==mat[i]]['extras'].sum(), d[d["match_id"]==mat[i]].reset_index()['start_date'][0])
      c+=1
    if c==5:
      break

  return summ/5
    
Team_data('Sunrisers Hyderabad','slow')
    
```
